## Task 7: Python Challenge (10 points)

Write a function called `is_even` that:
- Takes a number `n` as input
- Returns `True` if the number is even
- Returns `False` if the number is odd

Then test your function with at least 3 different numbers.

**Hint**: You'll need to learn about:
- How to define a function in Python (`def`)
- The modulo operator (`%`) - what does `n % 2` give you?
- `if`/`else` statements

_I approach this with the notion that the number must be an integer [(cf. Wikipedia on "parity")](https://en.wikipedia.org/wiki/Parity_(mathematics))._

````python
# Task 7: Write your is_even function here
import pandas as pd

def is_even(n):
    if type(n) == int:
        if n % 2 == 0:
            return True
        else:
            return False
    else:
        # Re returning strings, https://www.w3schools.com/python/python_functions.asp 
        return f"INVALID"

# Test it with at least 3 numbers
numbers = [1, 6, -8, 1.732, -22/7]
# Creating dictionary for neater display of results in pandas DataFrame
parity_dict = {}
for number in numbers:
    parity_dict[str(number)] = [is_even(number)]

parity_df = pd.DataFrame.from_dict(parity_dict, orient = "index", columns = ["Even?"])     # _df for DataFrame
display(parity_df)
````
